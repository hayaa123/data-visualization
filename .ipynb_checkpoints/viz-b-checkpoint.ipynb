```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

# Iris Data :
wich is a flowers data

```python
iris = sns.load_dataset("iris") 
```

finding some relation between data

```python
sns.scatterplot(x="sepal_length",y="sepal_width",data=iris,hue="species")
```

```python
sns.scatterplot(x="petal_length",y="petal_width",data=iris,hue="species")
```

from the figure above

1- we can determine which species is which by its petal dimensions 


```python
sns.scatterplot(x="petal_length",y="sepal_length",data=iris,hue="species")
```

from the figure above

we can determine wich flowe is which by its sepal and petal length

```python
sns.scatterplot(x="petal_width",y="sepal_width",data=iris,hue="species")
```

```python
sns.histplot(iris,x="species")
```

```python
count_dict = iris["species"].value_counts()
keys = count_dict.keys()
values= count_dict.values
plt.pie(values,labels=keys,autopct='%1.1f%%')
```

from the both figures above 

we can notice that the distribution if of data are uniform 

# conclusions 

1- we can determine which species is which by its petal dimensions 

2- we can determine which species is which by its betal and sepal length

3- the data are distributed in uniform distribution which mean that we have the same amount of sample on each species kind 

4- flowers in these three kind tend to increase in petal length lineary with betal width
